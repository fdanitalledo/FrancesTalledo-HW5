```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
url = 'https://raw.githubusercontent.com/AlR0d/courses/main/4421/utown.csv'
data = pd.read_csv(url)

# Question 1: Model 1
# price = 0 + 1 * sqft + 2 * utown + e

# a. Identify the base group (utown = 0, not near university)
base_group = data[data['utown'] == 0]
treatment_group = data[data['utown'] == 1]

# Prepare data for regression
X1 = data[['sqft', 'utown']]
X1 = sm.add_constant(X1)
y = data['price']

# Fit the model
model1 = sm.OLS(y, X1).fit()

# b. Plot the estimated sample regression line
sns.lmplot(x='sqft', y='price', hue='utown', data=data, palette={0: "blue", 1: "orange"}, markers=["o", "s"])
plt.title("Regression Line for Model 1")
plt.xlabel("Square Feet")
plt.ylabel("Price (in 1000s)")
plt.show()

# c. Interpret results
print("Model 1 Summary (Simplified):")
print(model1.summary().tables[1])

# Question 2: Model 2
# price = 0 + 1 * sqft + 2 * (utown * sqft) + e

# Prepare data for regression
data['utown_sqft'] = data['utown'] * data['sqft']
X2 = data[['sqft', 'utown_sqft']]
X2 = sm.add_constant(X2)

# Fit the model
model2 = sm.OLS(y, X2).fit()

# b. Plot the estimated sample regression line
sns.lmplot(x='sqft', y='price', hue='utown', data=data, palette={0: "blue", 1: "orange"}, markers=["o", "s"])
plt.title("Regression Line for Model 2")
plt.xlabel("Square Feet")
plt.ylabel("Price (in 1000s)")
plt.show()

# c. Interpret results
print("Model 2 Summary (Simplified):")
print(model2.summary().tables[1])

# Question 3: Model 3
# price = 0 + 1 * sqft + 2 * utown + 3 * (utown * sqft) + e

# Prepare data for regression
X3 = data[['sqft', 'utown', 'utown_sqft']]
X3 = sm.add_constant(X3)

# Fit the model
model3 = sm.OLS(y, X3).fit()

# b. Plot the estimated sample regression line
sns.lmplot(x='sqft', y='price', hue='utown', data=data, palette={0: "blue", 1: "orange"}, markers=["o", "s"])
plt.title("Regression Line for Model 3")
plt.xlabel("Square Feet")
plt.ylabel("Price (in 1000s)")
plt.show()

# c. Interpret results
print("Model 3 Summary (Simplified):")
print(model3.summary().tables[1])

# Question 4: Model 4 (adding remaining variables)
# price = 0 + 1 * sqft + 2 * utown + 3 * (utown * sqft) + 4 * age + 5 * pool + 6 * fplace + e

# Prepare data for regression
X4 = data[['sqft', 'utown', 'utown_sqft', 'age', 'pool', 'fplace']]
X4 = sm.add_constant(X4)

# Fit the model
model4 = sm.OLS(y, X4).fit()

# b. Interpret results
print("Model 4 Summary (Simplified):")
print(model4.summary().tables[1])

# Question 5: Model 5 (log-linear model)
# ln(price) = 0 + 1 * sqft + 2 * utown + 3 * (utown * sqft) + 4 * age + 5 * pool + 6 * fplace + e

# Transform price to log(price)
data['log_price'] = np.log(data['price'])

# Fit the model
model5 = sm.OLS(data['log_price'], X4).fit()

# a. Compute average percentage change due to pool
approx_change_pool = (np.exp(model5.params['pool']) - 1) * 100
exact_change_pool = model5.params['pool'] * 100

# b. Compute average percentage change in price for a 2,500-square-foot home near university
sqft_2500_near_uni = model5.params['const'] + model5.params['sqft'] * 25 + model5.params['utown'] + model5.params['utown_sqft'] * 25
sqft_2500_not_near_uni = model5.params['const'] + model5.params['sqft'] * 25
percentage_change_near_uni = (np.exp(sqft_2500_near_uni) - np.exp(sqft_2500_not_near_uni)) / np.exp(sqft_2500_not_near_uni) * 100

# c. Report Model 5 Summary
print("Model 5 Summary (Simplified):")
print(model5.summary().tables[1])
print(f"Average percentage change in price due to pool (approx): {approx_change_pool:.2f}%")
print(f"Average percentage change in price due to pool (exact): {exact_change_pool:.2f}%")
print(f"Average percentage change in price for 2,500 sqft home near university: {percentage_change_near_uni:.2f}%")

# Question 6: Report all models in one table
from statsmodels.iolib.summary2 import summary_col

models = [model1, model2, model3, model4, model5]
model_names = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5']

# Generate a summary table
summary_table = summary_col(models, stars=True, model_names=model_names, info_dict={'R-squared': lambda x: f"{x.rsquared:.2f}", 'Observations': lambda x: f"{int(x.nobs)}"})
print(summary_table)
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Load the data
url = 'https://raw.githubusercontent.com/AlR0d/courses/main/4421/beauty.csv'
data = pd.read_csv(url)

# Question 7: Are more people rated as above average looks or below average looks?
above_average_count = data['abavg'].sum()
below_average_count = data['belavg'].sum()
print(f"Above Average Count: {above_average_count}")
print(f"Below Average Count: {below_average_count}")

# Question 8: What percent of men are reported as having above average looks?
men_above_avg_percent = data[data['female'] == 0]['abavg'].mean() * 100
print(f"Percent of men with above average looks: {men_above_avg_percent:.2f}%")

# Question 9: What percent of women are reported as having above average looks?
women_above_avg_percent = data[data['female'] == 1]['abavg'].mean() * 100
print(f"Percent of women with above average looks: {women_above_avg_percent:.2f}%")

# Question 10: What is the base group of the regression in equation (5)?
print("The base group is individuals with average looks (neither above nor below average).")

# Question 11: Estimate equation (5) for both men and women
# ln(wage) = 0 + 1 * belavg + 2 * abavg + e
data['ln_wage'] = np.log(data['wage'])

# Separate by gender
men_data = data[data['female'] == 0]
women_data = data[data['female'] == 1]

# Fit model for men
X_men = men_data[['belavg', 'abavg']]
X_men = sm.add_constant(X_men)
model_men = sm.OLS(men_data['ln_wage'], X_men).fit()

# Fit model for women
X_women = women_data[['belavg', 'abavg']]
X_women = sm.add_constant(X_women)
model_women = sm.OLS(women_data['ln_wage'], X_women).fit()

# Display results
print("Model for Men (Equation 5):")
print(model_men.summary().tables[1])
print("\nModel for Women (Equation 5):")
print(model_women.summary().tables[1])

# Question 14: Pool men and women together into a single regression
# ln(wage) = 0 + 1 * belavg + 2 * abavg + 3 * female + 4 * (female * belavg) + 5 * (female * abavg) + e

# Create interaction terms
data['female_belavg'] = data['female'] * data['belavg']
data['female_abavg'] = data['female'] * data['abavg']

# Fit the pooled model
X_pooled = data[['belavg', 'abavg', 'female', 'female_belavg', 'female_abavg']]
X_pooled = sm.add_constant(X_pooled)
model_pooled = sm.OLS(data['ln_wage'], X_pooled).fit()

# Display pooled model results
print("Pooled Model (Equation 6):")
print(model_pooled.summary().tables[1])

# Interpret results for questions 12, 13, 15, and 16 based on the coefficients
# Question 12: Is there evidence that women with above average looks earn more than women with average looks?
women_abavg_coef = model_women.params['abavg']
print(f"Women with above average looks vs average: Coefficient for abavg (women) = {women_abavg_coef:.4f}")

# Question 13: Is there evidence that men with above average looks earn more than men with average looks?
men_abavg_coef = model_men.params['abavg']
print(f"Men with above average looks vs average: Coefficient for abavg (men) = {men_abavg_coef:.4f}")

# Question 15: Statistical difference between average looking males and females
female_coef = model_pooled.params['female']
print(f"Coefficient for female in pooled model (avg looking males vs females) = {female_coef:.4f}")

# Question 16: Statistical difference between above average looking females and below average looking females
female_abavg_coef = model_pooled.params['female_abavg']
print(f"Coefficient for female_abavg in pooled model (above avg females vs below avg females) = {female_abavg_coef:.4f}")
```
